```python
%matplotlib widget

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.stats import uniform
import scipy.integrate as integrate
from copy import deepcopy

plt.style.use('seaborn-v0_8')
```

```python
#Define the circe and the N=360 destinations
N = 360
destinations = np.linspace(0,2*np.pi,N)

#Assign a probability to each value
epsilon = 0.08
p_share = uniform.rvs(loc=0.5, scale=epsilon, size=N)
```

```python
# Compute expected utility
def Expected_Utility(ϕ, share_probs, β):
    approx_integral = np.mean(share_probs * np.sqrt(2-2*np.cos(ϕ-destinations)))
    return 1-((β/2)*approx_integral)

def E_U_of_circle(share_probs, β):
    return [Expected_Utility(θ, share_probs, β) for θ in destinations]
        
```

```python
# Seems reasonable
beta = 2.91
E_of_u_0 = E_U_of_circle(p_share, beta)

fig, axs = plt.subplots(1,2,figsize=(8,3))
axs[1].plot(destinations,E_of_u_0,lw=0.5,label=f'β = {beta}')
axs[0].scatter(destinations,p_share,lw=0.5,s=1,label=r'$\varepsilon = 0.08$')

axs[1].set_ylabel('Expected Utility [a.u]')
axs[1].set_xlabel(r'$\varphi$')
axs[1].legend()
axs[0].legend(loc='lower right')

axs[0].set_ylim([0.4,0.7])
axs[0].set_xlabel(r'$\varphi$')
axs[0].set_ylabel(r'$p_{share}$')
plt.tight_layout()
plt.savefig('Perturbed_Utility.png',format='png')
plt.show()
```

```python
plt.close()
```

```python
# Update the p_share values according to the propagator equation
def Update_pshare(pshare, expected_utility, Δt=0.05):
    new_pshare = pshare + Δt*pshare * (1-pshare) * expected_utility
    return new_pshare
```

```python
# Evolve the p-share values
beta = [1,np.pi/2,2,2.91,3]
Δt = 0.05
steps = 1000
Evol_of_pshare = np.empty((len(beta),steps,N)) #Each row a colection of p_share values at time t

for beta_index, β in enumerate(beta):
    new_p_share = deepcopy(p_share)
    expected_utilities = E_U_of_circle(new_p_share, β)
    
    for i in range(steps):
        new_p_share = Update_pshare(new_p_share, expected_utilities, Δt)
        expected_utilities = E_U_of_circle(new_p_share, β)
        Evol_of_pshare[beta_index,i,:] = new_p_share

    print(beta_index,' : ',β)
```

```python
# Create a single animation with multiple lines
fig, ax = plt.subplots(figsize=(7,5))
lines = [ax.plot([], [], lw=0.8, label=f'beta={β:.2f}')[0] for β in beta]
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(0, 1)
ax.set_xlabel(r'$\varphi$')
ax.set_ylabel(r'$p_{share}$')
ax.set_title(r'Evolution of $p_{share}$ Over Time for Different Beta Values')
ax.legend(loc='center left')

def init():
    for line in lines:
        line.set_data([], [])
    return lines

def update(frame):
    for beta_index, line in enumerate(lines):
        line.set_data(destinations, Evol_of_pshare[beta_index, frame])
    ax.set_title(f'Evolution of p_share Over Time (t: {frame})')
    return lines

ani = FuncAnimation(fig, update, frames=range(steps), init_func=init, blit=True, interval=50)

# Save the animation as a video file (uncomment to save)
ani.save('evolution_of_pshare_all_betas_1.gif', fps=30, writer='pillow')

# Display the animation
plt.show()
```
